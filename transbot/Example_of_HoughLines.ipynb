```python
# 霍夫曼直线检测的可视化
# 调用 sciplot 中的数学形式
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')

# 数据
dots = [(1,-1),(2,1),(3,3),(2,2),(3.3,1.5)]
# 画点图
plt.scatter(*zip(*dots))
plt.show()

# 构造多个函数 b = -xk + y，然后每个点构造一个函数，对交点进行加粗，交点涉及的线越多，交点越粗
for i in range(len(dots)):
    x = np.linspace(0,5,100)
    y = -dots[i][0]*x + dots[i][1]
    plt.plot(x,y)
# 统计交点和交点出现的次数，记录为字典 {(x,y):count}


plt.show()
```
